```python
import matplotlib.pyplot as plt
import pandas as pd
from windrose import WindroseAxes, plot_windrose

def wind_rose_plot(drct, sknt):

    ax = WindroseAxes.from_ax()
    ax.bar(drct, sknt, normed=True, opening=0.8, edgecolor='white')
    ax.set_legend()
    plt.show()

def windrose_from_df(df):

    winds = df[df.sknt > 0.0]
    #### Show a windrose with the speed/direction readings
    ws = pd.Series(winds['sknt'])
    wd = pd.Series(winds['drct'])
    wind_rose_plot(wd, ws)
```

```python
import time
print("Label with Hourly (rounded to nearest hour)")
# define an "Hourly" column
phx_df["Hourly"] = phx_df.index.round("1H")

# and a difference between each time and its rounded value
# (time2 - time1) => timedelta64.total_seconds()
print("Calculate OffsetFromHour as absolute difference from Hourly")
t0 = time.time()
phx_df["OffsetFromHour"] = ((phx_df.index -
                             phx_df["Hourly"])
                            .map(lambda td: td.total_seconds()/60)
                            .abs())

t = time.time()
print("That took %s seconds. Added as OffsetFromHour column." % (t-t0))

# Takes 10-15 minutes on mbai7, but cuts the number of readings by almost half
print("De-dup Hourly: Sort by OffsetFromHour, & keep only the one closest to Hourly")
t0 = time.time()
grouped = (phx_df.groupby("Hourly").
           apply(lambda by_hour: by_hour.sort_values(["OffsetFromHour"])).
           drop_duplicates(["Hourly"], keep="first"))
t = time.time()
print("That took %s seconds. Result saved in 'grouped'." % (t-t0))
plot_windrose(grouped, kind="bar", var_name="sknt", direction_name="drct",
              bins=[2,4,5,6,7,8,10,12,15,20])
plt.show()
# next: assign grouped to whatever you want to call these points going forward
```
