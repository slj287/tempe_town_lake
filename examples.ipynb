```python
import matplotlib.pyplot as plt
import pandas as pd
from windrose import WindroseAxes

def wind_rose_plot(drct, sknt):

    ax = WindroseAxes.from_ax()
    ax.bar(drct, sknt, normed=True, opening=0.8, edgecolor='white')
    ax.set_legend()
    plt.show()

def windrose_from_df(df):

    winds = df[df.sknt > 0.0]
    #### Show a windrose with the speed/direction readings
    ws = pd.Series(winds['sknt'])
    wd = pd.Series(winds['drct'])
    wind_rose_plot(wd, ws)
```

```python
import pandas as pd
from wind_analysis import get_deduped_df, by_month

grouped = get_deduped_df()
m2 = by_month(grouped)

monthly_avgs_before = [m2[mnum]['1995':'1998'].sknt.mean() for mnum in m2]
monthly_avgs_after_0 = [m2[mnum]['2000':'2003'].sknt.mean() for mnum in m2]
monthly_avgs_after_1 = [m2[mnum]['2004':'2007'].sknt.mean() for mnum in m2]
monthly_avgs_after_2 = [m2[mnum]['2011':'2014'].sknt.mean() for mnum in m2]
monthly_avgs_after_3 = [m2[mnum]['2014':'2017'].sknt.mean() for mnum in m2]
# Define a month name range that's open at the beginning
pin = pd.PeriodIndex(start='Dec', periods=13, freq="M").strftime("%b")

bva_dict = {
    "Before": monthly_avgs_before,
    "After (2000-2003)": monthly_avgs_after_0,
    "After (2004-2007)": monthly_avgs_after_1,
    "After (2011-2014)": monthly_avgs_after_2,
    "After (2014-2017)": monthly_avgs_after_3,
}
# We want the keys, but Before before the Afters
bva_order = list(bva_dict.keys())
bva_order.sort()
bva_order.insert(0, bva_order.pop())
bva_df = pd.DataFrame(bva_dict,
                      columns=bva_order, index=pin[1:13])
#bva_df.plot(use_index=True)
pins = pd.Series(list(range(0, 13)), index=pin)
bva_df.plot(xlim=(-1,12), xticks=pins)
plt.suptitle('Monthly average wind speed')
plt.show()
```

```python
from scipy.stats import chi2
import matplotlib.pyplot as plt
import numpy as np
from wind_analysis import get_deduped_df

grouped = get_deduped_df()
pre_grouped = grouped["1995":"1998"].sknt
grouped_by_month_pre = pre_grouped.groupby(pre_grouped.index.month)

def plot_month(m_df, ax, title, alpha=0.2):
    assert(not m_df.isna().any())

    dof = 5
    mean, var, skew, kurt = chi2.stats(dof, moments= 'mvsk')
    ax.set_xlim(-1,20)
    ax.set_ylim(0,.20)
    ax.set_title(title)
    
    x = np.linspace(chi2.ppf(0.01, dof),
                    chi2.ppf(0.99, dof), 100) 
    ax.plot(x, chi2.pdf(x, dof),
            'r-', lw=5, alpha=0.6, label='chi2 pdf')
    rv = chi2(dof)
    ax.plot(x, rv.pdf(x), 'k-', lw=2, label='pre-lake wind speeds')

    #ax.hist(m_df, density=True, histtype='stepfilled', alpha=0.2) 
    ax.hist(m_df, density=True) 
    ax.legend(loc='best', frameon=False)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
plot_month(grouped_by_month_pre.get_group(1), axes[0], "January")
plot_month(grouped_by_month_pre.get_group(7), axes[1], "July")
plt.show()
```

```python
from scipy.stats import chi2
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
from wind_analysis import get_deduped_df

grouped = get_deduped_df()
pre_grouped = grouped["1995":"1998"].sknt.dropna()
grouped_by_month_pre = pre_grouped.groupby(pre_grouped.index.month)

norm = stats.norm(0, 1)

def plot_month_v0(m_df, ax, title):
    #f = grouped_by_month_pre.get_group(1)
    ax.set_xlim(-1,20)
    ax.set_ylim(0,.20)
    ax.set_title(title)

    #plt.hist(f, bins=list(range(0, 30, 3)), normed=True)
    ax.hist(m_df, normed=True)
    #plt.hist(f, bins = 'auto')

    # Plot the theoretical density of f
    x = np.arange(0, 20, .05)
    ax.plot(x, stats.chi2.pdf(x, df=6), color='r', lw=2)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
plot_month(grouped_by_month_pre.get_group(1), axes[0], "January (same as above)")
plot_month_v0(grouped_by_month_pre.get_group(1), axes[1], "January (alternative $\chi^2$)")

plt.show()
```
