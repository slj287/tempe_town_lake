```python
import matplotlib.pyplot as plt
import pandas as pd
from windrose import WindroseAxes

def wind_rose_plot(drct, sknt):

    ax = WindroseAxes.from_ax()
    ax.bar(drct, sknt, normed=True, opening=0.8, edgecolor='white')
    ax.set_legend()
    plt.show()

def windrose_from_df(df):

    winds = df[df.sknt > 0.0]
    #### Show a windrose with the speed/direction readings
    ws = pd.Series(winds['sknt'])
    wd = pd.Series(winds['drct'])
    wind_rose_plot(wd, ws)
```

```python
windrose_from_df(phx_df)
```

```python
windrose_from_df(phx_df[phx_df.index.year < 1999])
```

```python
# For grouping by month, it would be nice to know what month number
# (for example) "Nov" is.  To do so, we enumerate the months starting
# from 1 (for January), and build a dict that can be used for looking
# up the number for any given month.
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
month_numbers = dict([(month, number)
                      for number, month in enumerate(months, 1)])
month_dfs = {}
for month_number, month_name in enumerate(months, 1):
    month_dfs[month_name] = phx_df[phx_df.index.month == month_number]
```
